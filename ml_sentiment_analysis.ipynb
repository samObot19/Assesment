```python
import pandas as pd

file_path = '/content/drive/MyDrive/datasets/twitter_training.csv'

# Load dataset
df = pd.read_csv(file_path, header=None, names=["ID", "Topic", "Sentiment", "Tweet"])
df["Tweet"] = df["Tweet"].str.strip()

df.head()
```

# Drop unnecessary columns (like ID and Topic)


```python
df = df.drop(columns=["ID", "Topic"])
print(df.columns)
```

# Text preprocessing

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab') # Download the missing 'punkt_tab' data

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # Convert the input to string to handle potential float values
    text = str(text)
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)  # remove URLs
    text = re.sub(r'\@w+|\#', '', text)  # remove @mentions and hashtags
    text = re.sub(r"[^a-zA-Z]", " ", text)  # remove numbers/symbols
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words and len(w) > 2]
    return " ".join(tokens)

df['Cleaned_Tweet'] = df['Tweet'].apply(preprocess)
```
